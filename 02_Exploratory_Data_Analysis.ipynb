Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Load the Dataset

```python
df = pd.read_csv("bank_customer_cleaned.csv")
```

Churn Distribution

```python
churn = df['Exited'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    churn,
    labels=['Stayed','Exited'],
    autopct='%1.1f%%'
)

plt.title("Customer Churn Distribution")

plt.show()
```

Gender Distribution

```python
gender = df['Gender'].value_counts()

plt.figure(figsize=(6,5))

gender.plot(kind='bar')

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Customers")

plt.show()
```

Geography Distribution

```python
geo = df['Geography'].value_counts()

plt.figure(figsize=(8,5))

geo.plot(kind='bar')

plt.title("Customers by Geography")
plt.xlabel("Country")
plt.ylabel("Customers")

plt.show()
```

Age Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df['Age'], bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

Credit Score Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df['CreditScore'], bins=20)

plt.title("Credit Score Distribution")
plt.xlabel("Credit Score")
plt.ylabel("Customers")

plt.show()
```

Balance Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df['Balance'], bins=20)

plt.title("Balance Distribution")
plt.xlabel("Balance")
plt.ylabel("Customers")

plt.show()
```

Salary Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df['EstimatedSalary'], bins=20)

plt.title("Estimated Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Customers")

plt.show()
```

Churn by Geography

```python
pd.crosstab(
    df['Geography'],
    df['Exited']
).plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Customer Churn by Geography")
plt.xlabel("Country")
plt.ylabel("Customers")

plt.show()
```

Churn by Gender

```python
pd.crosstab(
    df['Gender'],
    df['Exited']
).plot(
    kind='bar'
)

plt.title("Customer Churn by Gender")
plt.xlabel("Gender")
plt.ylabel("Customers")

plt.show()
```

Active Members vs Churn

```python
pd.crosstab(
    df['IsActiveMember'],
    df['Exited']
).plot(
    kind='bar'
)

plt.title("Active Members vs Churn")
plt.xlabel("Active Member")
plt.ylabel("Customers")

plt.show()
```

Products vs Churn

```python
pd.crosstab(
    df['NumOfProducts'],
    df['Exited']
).plot(
    kind='bar'
)

plt.title("Products vs Customer Churn")
plt.xlabel("Number of Products")
plt.ylabel("Customers")

plt.show()
```

Credit Card vs Churn

```python
pd.crosstab(
    df['HasCrCard'],
    df['Exited']
).plot(
    kind='bar'
)

plt.title("Credit Card Ownership vs Churn")
plt.xlabel("Has Credit Card")
plt.ylabel("Customers")

plt.show()
```

Correlation Matrix

```python
numeric_df = df.select_dtypes(include=['number'])

corr = numeric_df.corr()

plt.figure(figsize=(12,8))

plt.imshow(corr, cmap='coolwarm', aspect='auto')
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")

plt.show()
```
